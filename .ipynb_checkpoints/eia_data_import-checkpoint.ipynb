```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from config import api_key
import eia
import re
from sklearn.linear_model import LinearRegression
```

```python
#save variable to call in api data w/ key
api = eia.API(api_key)
```

```python
#save all relevant api endings as an iterable object
series=['TOTAL.TEICBUS.M','TOTAL.TERCBUS.M','TOTAL.TEACBUS.M','TOTAL.TXCCBUS.M','TOTAL.TXEIBUS.M','TOTAL.TXICBUS.M','TOTAL.TXRCBUS.M','TOTAL.TXACBUS.M','TOTAL.TETCBUS.M']
```

```python
#test one api ending; import to pandas dataframe
newdf=api.data_by_series(series='TOTAL.TECCBUS.M')
dataframe1=pd.DataFrame(newdf)
dataframe1.head()
```

```python
import datetime
from datetime import date
```

```python
datetime.date(2011, 3, 1)
```

```python
#iterate through api endings and save to above dataframe
for x in series:
    series_search = api.data_by_series(series=x)
    df = pd.DataFrame(series_search)
    dataframe1=dataframe1.join(df, how="outer")
```

```python
#print all data in df
dataframe1.head()
```

```python
#add index (unique identifier)
dataframe1=dataframe1.reset_index()
dataframe1=dataframe1.rename(columns={"index":"Time"})
dataframe1.head()
```

```python
#call on first item using index 
dataframe1["Time"][0]
```

```python
#split time calumn into separate year and month (test on first line)
year = re.findall('....', (dataframe1['Time'][0]))[0]
month = re.findall('....', (dataframe1['Time'][0]))[1]

print(f"year = {year}")
print(f"month = {month}")
```

```python
#create empty lists for future month and year columns
year = []
month = []
counter = 0
```

```python
#iterate to split all time columns into year/month; add to respective list
for time in dataframe1['Time']:
    year.append(re.findall('....', (dataframe1['Time'][counter]))[0])
    month.append(re.findall('....', (dataframe1['Time'][counter]))[1])
    counter = counter + 1
```

```python
#add month and year columns to df
dataframe1['Year'] = year
dataframe1['Month'] = month
```

```python
# use this parent dataframe to pull relevant info into children dataframes for analysis
dataframe1.head()
```

```python
timestamp = dataframe1["Time"]
datetime_list = []
```

```python
# fill empty list with dates converted to datetime
for i in timestamp:
    i = i.replace(" ", "-")
    i = i + "01"
    i = pd.to_datetime(i, infer_datetime_format = True)
    datetime_list.append(i)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace_total = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumption, Monthly (Trillion Btu)'],
                  name='Total Consumption',
                  marker=dict(color='red'))

trace_electric = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumed by the Electric Power Sector, Monthly (Trillion Btu)'],
                  name='Consumption by Electricity Generation',
                  marker=dict(color='blue'))

trace_commercial = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumed by the Commercial Sector, Monthly (Trillion Btu)'],
                  name='Consumption by Commercial Sector',
                  marker=dict(color='green'))

trace_industrial = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumed by the Industrial Sector, Monthly (Trillion Btu)'],
                  name='Consumption by Industrial Sector',
                  marker=dict(color='yellow'))

trace_residential = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumed by the Residential Sector, Monthly (Trillion Btu)'],
                  name='Consumption by Residential Sector',
                  marker=dict(color='orange'))

trace_transportation = go.Scatter(x=primary_consumption_datestamp_df['Date'],
                  y=primary_consumption_datestamp_df['Total Primary Energy Consumed by the Transportation Sector, Monthly (Trillion Btu)'],
                  name='Consumption by Transportation Sector',
                  marker=dict(color='orange'))

data = [trace_total, trace_electric, trace_commercial, trace_industrial, trace_residential, trace_transportation]

layout = go.Layout(title="Primary Energy Consumption by Sector",
                xaxis=dict(title='Time'),
                yaxis=dict(title='Energy Consumption in Trillion British Thermal Units(BTU)'))

fig = go.Figure(data=data, layout=layout)

url = py.plot(fig, filename='consumption-by-sector')
```
